Correlation with the Pandas library can be done using the **corr()** method, by applying the method parameter for 'method=pearson' or 'method=spearman'
- The Pearson correlation measures the **linear** relationship between two variables.
- The Spearman correlation measures the **the monotonic association** between two variables.

```python
#pairwise correlation of all column in a DF
correlation_matrix = df.corr()

#correlation on a specific column with all the others in DF
correlation_of_A = df['A'].corr(df)

#correlation between 2 specific columns of a DF
correlation_AB = df['A'].corr(df['B'])

#correlation betwee 2 Dataframes
correlation_df1_df2 = df1.corrwith(df2)

#applying different types of correlation
correlation_matrix_spearman = df.corr(method='spearman')
#also available: pearson and kendall
```

<p style="background-color:#F0EACF; font-size:16px;padding:15px;color:#03045eff; font-family: normal; border-radius:10px">
SCIPY.STATS: with the scipy.stats.pearsonr() & scipy.stats.spearmanr() function
</p>

```python
from scipy.stats import pearsonr

x = df['A'].values
y = df['B'].values

corr, p_value = pearsonr(x, y)
print(corr, p_value)


#OR
from scipy.stats import spearmanr

x = df['A'].values
y = df['B'].values

corr, p_value = spearmanr(x, y)
print(corr, p_value)
```

SEABORN
</p>

The seaborn library is built on top of maptlotlib and provides a good way to use correlation visualizations with the two most common being pairplots and heatmap.

Both pairplots and heatmap allow you to identify patterns in your data - the strength and direction of any correlation.

```python
#heatmap
import seaborn as sns

corr = df.corr()
sns.heatmap(corr)

#OR
sns.heatmap(corr, annot=True, cmap='coolwarm')
#annot provides the coefficient values on a heatmap
#cmap provides the map color
```

```python
#pairplot
sns.pairplot(df)

#OR
sns.pairplot(df, kind='reg', hue='species')
#the 'kind' parameter specifies the type of plot 
#the 'hue' provides a color schema
```
